```python
data = data.sort_values(by='avg', ascending=False)
data
```

```python
def highlight_max(s):
    if s.dtype == object or s.dtype == np.float64:
        is_max = [False for _ in range(s.shape[0])]
    else:
        is_max = s == s.max()
    return ['background: lightgreen' if cell else '' for cell in is_max]
  
data.style.apply(highlight_max)
```

```python
boys = data[data['Gender']=="Male"]
#boys

def highlight_max(s):
    if s.dtype == object or s.dtype == np.float64:
        is_max = [False for _ in range(s.shape[0])]
    else:
        is_max = s == s.max()
    return ['background: lightgreen' if cell else '' for cell in is_max]
  
boys.style.apply(highlight_max)
```

```python
girls = data[data['Gender']=="Female"]
#girls

def highlight_max(s):
    if s.dtype == object or s.dtype == np.float64:
        is_max = [False for _ in range(s.shape[0])]
    else:
        is_max = s == s.max()
    return ['background: lightgreen' if cell else '' for cell in is_max]
  
girls.style.apply(highlight_max)
```
